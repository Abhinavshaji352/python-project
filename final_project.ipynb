```python
import plotly.express as px

ten_parties_names = df['party'].value_counts().head(10).index

# Filter the DataFrame to include only the top 10 parties
df_top_parties = df[df['party'].isin(ten_parties_names)]
region_sales = df_top_parties.groupby('party')['position'].sum().reset_index()
fig = px.bar(region_sales,
             x='party',
             y='position',
             color='party',
             title='Total Position by Top 10 Parties')

fig.show()
```

```python
#multivarie Analysis

top_10_party_names = df['party'].value_counts().head(10).index.tolist()
df_filtered_parties = df[df['party'].isin(top_10_party_names)].copy()

df_filtered_parties['vote_share_percentage'] = pd.to_numeric(df_filtered_parties['vote_share_percentage'], errors='coerce')
df_filtered_parties['vote_share_percentage'].fillna(df_filtered_parties['vote_share_percentage'].mean(), inplace=True)

df_filtered_parties['position_cat'] = df_filtered_parties['position'].astype('category')

plt.figure(figsize=(12, 6))
sns.barplot(data=df_filtered_parties, x='party', y='vote_share_percentage', hue='position_cat', errorbar=None)
plt.title('Average Vote Share Percentage by Top Parties and Position')
plt.xlabel('Party')
plt.ylabel('Average Vote Share Percentage')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Position')
plt.tight_layout()
plt.show()
```

```python
# Create a pivot table to summarize average vote share percentage

pivot_table_avg_vote_share = df_filtered_parties.pivot_table(
    index='party',
    columns='position_cat',
    values='vote_share_percentage',
    aggfunc='mean'
)

print("Pivot Table of Average Vote Share Percentage:")
display(pivot_table_avg_vote_share.head())
```
